```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score, accuracy_score, precision_recall_curve, precision_score, recall_score,  confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the data
pd.set_option("display.max_columns",None) 
data = pd.read_csv(r"C:\Users\91892\OneDrive\Documents\CCDP_docs\UCI_Credit_Card.csv")
print("Shape of the data", data.shape)
data.head(5)
```

```python
X=data.copy(deep = True)
Y=data['default.payment.next.month']
X.drop(['ID','default.payment.next.month'], axis=1, inplace = True)
```

```python
x_train,x_valid,y_train,y_valid = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
def scores(x_train, y_train, x_valid, y_valid):
    # model 
    model_RFC = RandomForestClassifier()
    model_LR = LogisticRegression()
    model_GBR = GradientBoostingClassifier()
    model_DTC = DecisionTreeClassifier()
    model_KN = KNeighborsClassifier()
    # fitting the model
    model_RFC.fit(x_train, y_train)
    model_LR.fit(x_train, y_train)
    model_GBR.fit(x_train, y_train)
    model_DTC.fit(x_train, y_train)
    model_KN.fit(x_train, y_train)
    # predict the model
    y_RFC=model_RFC.predict(x_valid)
    y_LR=model_LR.predict(x_valid)
    y_GBR=model_GBR.predict(x_valid)
    y_DTC=model_DTC.predict(x_valid)
    y_KN=model_KN.predict(x_valid)
    #score
    print("F1 Score")
    f1_RFC = f1_score(y_valid, y_RFC)
    f1_LR = f1_score(y_valid,y_LR)
    f1_GBR = f1_score(y_valid, y_GBR)
    f1_DTC = f1_score(y_valid,y_DTC)
    f1_KN = f1_score(y_valid,y_KN)
    print(("RFC : {}; LR : {}; GBR : {}; DTC : {}; KN : {}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))
    
    print("accuracy_score")
    f1_RFC = accuracy_score(y_valid, y_RFC)
    f1_LR = accuracy_score(y_valid,y_LR)
    f1_GBR = accuracy_score(y_valid, y_GBR)
    f1_DTC = accuracy_score(y_valid,y_DTC)
    f1_KN = accuracy_score(y_valid,y_KN)
    print(("RFC : {}; LR : {}; GBR : {}; DTC : {}; KN : {}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))

    print("precision_score")
    f1_RFC = precision_score(y_valid, y_RFC)
    f1_LR = precision_score(y_valid,y_LR)
    f1_GBR = precision_score(y_valid, y_GBR)
    f1_DTC = precision_score(y_valid,y_DTC)
    f1_KN = precision_score(y_valid,y_KN)
    print(("RFC : {}; LR : {}; GBR : {}; DTC : {}; KN : {}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))

    print("recall_score")
    f1_RFC = recall_score(y_valid, y_RFC)
    f1_LR = recall_score(y_valid,y_LR)
    f1_GBR = recall_score(y_valid, y_GBR)
    f1_DTC = recall_score(y_valid,y_DTC)
    f1_KN = recall_score(y_valid,y_KN)
    print(("RFC : {}; LR : {}; GBR : {}; DTC : {}; KN : {}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))
    
    print("precision_recall_curve")
    f1_RFC = precision_recall_curve(y_valid, y_RFC)
    f1_LR = precision_recall_curve(y_valid,y_LR)
    f1_GBR = precision_recall_curve(y_valid, y_GBR)
    f1_DTC = precision_recall_curve(y_valid,y_DTC)
    f1_KN = precision_recall_curve(y_valid,y_KN)
    print(("RFC :\n{}; LR :\n{}; GBR :\n{}; DTC :\n{}; KN :\n{}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))
    
    print("confusion_matrix")
    f1_RFC = confusion_matrix(y_valid, y_RFC)
    f1_LR = confusion_matrix(y_valid,y_LR)
    f1_GBR = confusion_matrix(y_valid, y_GBR)
    f1_DTC = confusion_matrix(y_valid,y_DTC)
    f1_KN = confusion_matrix(y_valid,y_KN)
    print(("RFC :\n{};\nLR :\n{};\nGBR :\n{};\nDTC :\n{};\nKN :\n{}").format(f1_RFC,f1_LR,f1_GBR,f1_DTC,f1_KN))

    #precision_recall_curve
```

```python
scores(x_train, y_train, x_valid, y_valid)
```
